```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point):
    def deriv(f, base_point):
        # Estimate the derivative of the function f at base_point using the symmetric approximation
        return (f(base_point + 10**(-10)) - f(base_point - 10**(-10))) / (2 * 10**(-10))
    
    x_coords = [initial_point]  # This list is where you will store the x_n's
    y_coords = [f(initial_point)]  # This list is where you will store the y_n's
    
    for _ in range(10000):  # Limit the number of steps to 10000
        current_x = x_coords[-1]
        current_derivative = deriv(f, current_x)
        
        if abs(current_derivative) < 1e-10:  # Stop if the derivative is close to 0
            break
        
        next_x = current_x - learning_rate * current_derivative
        x_coords.append(next_x)
        y_coords.append(f(next_x))
    
    # Plotting portion. 
    plot_range = np.linspace(min(x_coords) - 0.5, max(x_coords) + 0.5, 10000)  # A nice plot range to make it look good
    function_range = [f(i) for i in plot_range]
    
    plt.plot(plot_range, function_range)  # This plots the function f(x)
    plt.plot(x_coords, y_coords, marker='o')  # This will plot the sequence of points x_n, f(x_n)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Gradient Descent')
    plt.show()
    
    return round(x_coords[-1], 3), round(y_coords[-1], 3)  # Returns your last x_n and y_n, rounded to three decimal places

# Example usage:
def example_function(x):
    return x**2

# Finding the local minimum of f(x) = x^2 with a learning rate of 0.8 and initial point of 1
min_x, min_y = gradient_descent(example_function, learning_rate=0.8, initial_point=1)
print(f"Local minimum at x = {min_x}, f(x) = {min_y}")
 
```

```python
import matplotlib.pyplot as plt
import numpy as np

def deriv(f, base):
    return (f(base + 10**(-10)) - f(base - 10**(-10))) / (2 * 10**(-10))

def gradient_descent(f, L, x0):
    x_coords = []
    y_coords = []
    x_coords.append(x0)  # Set the first x coordinate to be x0
    y_coords.append(f(x0))  # Set the first y coordinate to be f(x0)
    xi = x0
    i = 1
    while abs(deriv(f, xi)) > 0.01:
        xi = x_coords[i-1] - L * deriv(f, x_coords[i-1])
        yi = f(xi)
        x_coords.append(xi)  # Add xi to x as the last value
        y_coords.append(yi)  # Add yi to y as the last value
        i += 1
    
    return x_coords, y_coords

def funny_function(x):
    if x > 0:
        return x**x
    elif x == 0:
        return 1
    else:
        return abs(x)**abs(x)

# Gradient descent for positive initial point
x_coords_pos, y_coords_pos = gradient_descent(funny_function, 0.87, 1)

# Gradient descent for negative initial point
x_coords_neg, y_coords_neg = gradient_descent(funny_function, 0.87, -1)

# Plotting portion with x-axis range from -1.5 to 1.5
plot_range = np.linspace(-1.5, 1.5, 10000)
function_range = [funny_function(i) for i in plot_range]

plt.plot(plot_range, function_range)
plt.plot(x_coords_pos, y_coords_pos, marker='o', label='Positive Initial Point')
plt.plot(x_coords_neg, y_coords_neg, marker='o', label='Negative Initial Point')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent')
plt.xlim(-1.5, 1.5)  # Set x-axis limits
plt.show()

print(f"Positive Initial Point: Local minimum at x = {round(x_coords_pos[-1], 3)}, f(x) = {round(y_coords_pos[-1], 3)}")
print(f"Negative Initial Point: Local minimum at x = {round(x_coords_neg[-1], 3)}, f(x) = {round(y_coords_neg[-1], 3)}")
```
